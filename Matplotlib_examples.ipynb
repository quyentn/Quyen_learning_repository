### I. Creating and Customizing Our First Plots

___KEY CONCEPTS___:
- Plots are constructed with at least 2 python lists for x and y coordinators

__Methods:__
- ___matplotlib.pyplot.plot___ (plt.plot) for line graphs
- ___plt.show()___ at the end of all plt python codes for plot visualized
- ___multiple plt.plot___ before plt.show() for displaying multiple plots within a graph. Plot in later order will overlap the previous one
- ___plt.title___ for graph title on top of the plot
- ___plt.xlabel/plt.ylabel___ for plot labels in the axes
- ___plt.legend([list])___ for displaying legends based on orders of plots
- ___plt.legend()___ for displaying legends based on label variable in each plot
- ___plt.tight$\_$layout___ for fitting graph in a tight screen like laptop
- ___plt.grid(True)___ for allowing grid lines (default = False)
- ___plt.style.use(style)___ for applying a specific style to the plot. Use print(plt.style.available) to see available styles
- ___plt.xkcd()___ for distorting the final visualization (defalt = True --> make it False inside parentheses to turn off) 
- ___plt.savefig('plotName.png')___ for saving plot to a png file

__Variables inside plt.plot(and or other plot types):__
- ___fmt =___ '[marker][line][color]' for formatting of a single plot inside the graph (fmt = can be omitted)
- ___color =___ (eg: 'k' for black OR 6-char hash colors) for selecting color of a single plot (default = blue or 'b')
- ___linestyle =___ (eg: '--' for dashed line) for selecting line style of a single plot (default = solid line)
- ___marker =___ (eg: '.' for dotted marker) for selecting marker style of a single plot (default = None)
- ___linewidth =___ (eg: 3 for triple bold line) for selecting line width of a single plot (default = 1)
- ___label =___ [label name] for selecting label of a single plot (default = None)

### II. Bar Charts and Analyzing Data from CSVs

___KEY CONCEPTS:___

___Methods:___
- ___plt.bar___ method instead of plot for bar chart
- ___plt.barh___ for horizontal bar plot
- ___plt.bar___ and ___plt.barh___ in python2.7 do not take string list as input. We may need to take indexes as interger input instead and then display labels for those indexes using plt.xticks/plt.yticks
- ___Add more bar charts___ by putting more plt.bar after each other. But they will overlap each other
- ___.reverse()___ method at the end of each list to reserve the list order if necessary
- ___plt.xticks(x$\_$pos, tuple(x$\_$list))___ to display different labels from x_list to numpy indexes in x_pos
- ___Counter().update___(row.split(';')) to update counter dictionary to count elements row by row. Counter result is a dict but appears like a list of multiple tuples with 2 elements of text and count value.
- ___Counter().most$\_$common(int)___ for selecting top count records only

___Variables:___
- ___labels =___  and ___colors =___ are similar to those of line plot
- ___width =___ (eg. 0.25) for adjusting bar width and shifting bars next to each other in one x position (default = 0.8)
- ___np.arange(len(x$\_$list))___ for creating numpy array indexes to put in position of x in plot function

```python
languages = []
popularity = []
for item in language_counter.most_common(15):
    languages.append(item[0])
    popularity.append(item[1])
print(languages)
print(popularity)
```

### III. Pie Charts

___KEY CONCEPTS:___

___Methods:___
- ___plt.pie___ method to create pie chart. Pie chart take 1 list of numeric values for slice percentages


___Variables:___
- ___labels =___  and ___colors =___ are similar to those of line plot
- ___wedgeprops={}___ (eg. {'edgecolor':'black'}) for wedge properties like edgecolor,...
- ___explode = []___ for array of numeric radius fraction away from pie midpoint that we want a slice to be exploded 
- ___shadow = True/False___ for allowing shadow (default = False)
- ___startangle =___ (eg. 90 for 90 degree or vertical line) for starting angle of the first slice in the list
- ___autopct=___ (eg. '%1.1f%%' for format of percentage with 1 decimal position) for allowing percentage labels on top of each slice

Follow this video: https://youtu.be/MPiz50TsyF0
- plt.pie with 1 list as input data
- wedgeprops to edit edge color and other wed properties

```python
from matplotlib import pyplot as plt
plt.style.use('fivethirtyeight')
# Language Popularity
slices = [120,80,30,20] #number don't need to add up to 100
labels = ['Sixty','Forty','Extra','Extra2']
#colors = ['blue','red','purple','yellow']
colors = ['#008fd5','#fc4f30','#7f38ec','#e5ae37']
plt.pie(slices,labels = labels,colors=colors,wedgeprops={'edgecolor':'black'})
plt.title('My Awesome Pie Chart')
plt.tight_layout()
plt.show()
```

```python
from matplotlib import pyplot as plt
plt.style.use('fivethirtyeight')
# Language Popularity
slices = [59219, 55466, 47544, 36443, 35917, 31991, 27097, 23030, 
          20524, 18523, 18017, 7920, 7331, 7201, 5833]
labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java', 'Bash/Shell/PowerShell', 'C#', 'PHP', 
          'C++', 'TypeScript', 'C', 'Other(s):', 'Ruby', 'Go', 'Assembly']
explode = [0,0,0,0.1,0]
#build chart of top 5 only
plt.pie(slices[:5],labels = labels[:5], explode = explode, shadow = True, 
        startangle = 90,autopct='%1.1f%%',
        wedgeprops={'edgecolor':'black'})

plt.title('My Awesome Pie Chart')
plt.tight_layout()
plt.show()
```

### IV. Stack Plots

___KEY CONCEPTS:___ Stack plot is also called area plot

___Methods:___
- ___plt.stackplot___ method to create stacked chart of multiple charts/players. The first element in stackplot is still x$\_$list, all the next input lists will be stacked together into y axis
- __plt.legend(loc=loc)__ (eg. loc = 'upper left'/'lower left') for allowing legend and specifying location for legend

___Variables:___
- ___labels =___  and ___colors =___ are similar to those of line plot but label len and color len will be the number of stacked players, not the len of x$\_$list
- ___loc=___ inside legend (see above). We can specify coordinates for the legend position using of tuple of (x,y)

### V. Filling Area on Line Plots

___KEY CONCEPTS:___

___Methods:___
- ___plt.fill$\_$between___ method to create fill some shaded color between 2 lines (either line chart line or fixed value horizontal line). 
- The first input in fill$\_$between is still x$\_$list. 
- The next input can be 1 or 2 lines (when only 1 line is specified, the plot will default to fill between the given line and y=0 horizontal line). 
- This plot can take maxium of 2 lines after x value (total 3 numeric input lists).

___Variables:___
- ___alpha =___ (max = 1, min =0) for the degree of shaded color see-through
- ___where = (condition)___ for adding coditional formating to only shade some parts of the plot
- ___interpolate =___ (eg. True/False) True for making sure that certain x intersection won't get clipped 
- ___labels =___  and ___colors =___ are similar to those of line plot for customizing color and label for a specific shaded area.

```python
import pandas as pd
from matplotlib import pyplot as plt
import io
import requests
import warnings
warnings.simplefilter("ignore")
url = 'https://raw.githubusercontent.com/CoreyMSchafer/code_snippets/master/Python/Matplotlib/05-Fill_Betweens/data.csv'
s = requests.get(url,verify=False).content
data = pd.read_csv(io.StringIO(s.decode('utf-8')))
data.head(5)
```

- ___plt.gcf()___ method for getting current figure. We can change sub method/ variable of current figure like x axis format
- ___plt.gca()___ method for getting current axis, then we can change axis format using defined formatter
- ___mpl$\_$dates.DateFormatter(fmt$\_$style)___ (eg:'%b, %d %Y') for defining date formatter. See reference below (https://docs.python.org/3/library/datetime.html#strftime-and-strptime-behavior)

___Variables:___

- ___plt.gcf().autofmt$\_$xdate()___ for automatically making date label in x axis nice looking (like inclined text)
- ___plt.gca().xaxis.set$\_$major$\_$formatter(fmt$\_$style)___ (eg. mpl$\_$dates.DateFormatter(...)) for defining specific date style format
- ___inplace=True___ for adding in pandas sort_values method to overwrite the original data without doing expression like data = data.someMethod...

```python
import pandas as pd
from matplotlib import pyplot as plt
from datetime import datetime, timedelta
from matplotlib import dates as mpl_dates
import io
import requests
import warnings
warnings.simplefilter("ignore")
url = 'https://raw.githubusercontent.com/CoreyMSchafer/code_snippets/master/Python/Matplotlib/08-TimeSeries/data.csv'
s = requests.get(url,verify=False).content
data = pd.read_csv(io.StringIO(s.decode('utf-8')))
data.head(3)
```
